## Limits on Heavy Neutrinos

This notebook plots the standard parameter space of heavy neutrinos under the assumption of single flavor dominance.

The data is pulled from [this spreadsheet](https://docs.google.com/spreadsheets/d/1p_fslIlThKMOThGl4leporUsogq9TmgXwILntUZOscg/edit?usp=sharing), and limits are drawn from digitized or shared data.



```python
%load_ext autoreload
%autoreload 2
```

```python
import numpy as np
import pandas as pd
from matplotlib.pyplot import *
from Nlimits import plot_tools
from Nlimits import hnl_tools
```

```python
case_electron = hnl_tools.Limits(flavor='e', invisible=True)
case_muon = hnl_tools.Limits(flavor='mu', invisible=True)
case_tau = hnl_tools.Limits(flavor='tau', invisible=True)
```

#### Start by printing the combination of all lines to file

```python
x=np.geomspace(1,5e4)

np.save("combined_limits/Ue4_inv", np.array([x,case_electron.interp_func_all(x)]))
np.savetxt("combined_limits/Ue4_inv.dat", np.array([x,case_electron.interp_func_all(x)]).T)
x,y = np.load("combined_limits/Ue4_inv.npy")
x,y = np.loadtxt("combined_limits/Ue4_inv.dat", unpack=True)

np.save("combined_limits/Umu4_inv", np.array([x,case_muon.interp_func_all(x)]))
np.savetxt("combined_limits/Umu4_inv.dat", np.array([x,case_muon.interp_func_all(x)]).T)
x,y = np.load("combined_limits/Umu4_inv.npy")
x,y = np.loadtxt("combined_limits/Umu4_inv.dat", unpack=True)

np.save("combined_limits/Utau4_inv", np.array([x,case_tau.interp_func_all(x)]))
np.savetxt("combined_limits/Utau4_inv.dat", np.array([x,case_tau.interp_func_all(x)]).T)
x,y = np.load("combined_limits/Utau4_inv.npy")
x,y = np.loadtxt("combined_limits/Utau4_inv.dat", unpack=True)
```
